```python
# Cria uma lista com as moedas de interesse
moedas_interesse = ["bitcoin", "ethereum"]
```

```python
# Busca os dados
dados = []
for moeda in moedas_interesse:
  resposta = requests.get(f"{API_BASE_URL}/simple/price?vs_currencies=usd&ids={moeda}")
  dados.append(resposta.json())
```

```python
dados
```

```python
# Prepara os dados para posterior análise
dados_coletados = {
    "moedas": [],
    "valor": [],
    "data_coleta": []
}
```

```python
# Bibliotecas auxiliares
from time import sleep
from datetime import datetime
```

```python
# Coleta os dados
for i in range(10):
  for moeda in moedas_interesse:
    resposta = requests.get(f"{API_BASE_URL}/simple/price?vs_currencies=usd&ids={moeda}")
    print(resposta.json())
    # Coloca os dados da resposta em dados_coletados
    dados_coletados["moedas"].append(moeda)
    dados_coletados["valor"].append(resposta.json()[moeda]["usd"])
    dados_coletados["data_coleta"].append(datetime.now())
  sleep(10)

```

```python
dados_coletados
```

```python
# Convertendo os dados para dataframe
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(dados_coletados)
```

```python
# Salva os dados em um JSON
df.to_json("dados_moedas.json")
```

```python
# Salva os dados em um CSV
df.to_csv("dados_moedas.csv")
```

```python
import requests
from bs4 import BeautifulSoup

# 1. Buscar o conteúdo do site
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)
print(page.text)  # Exibe todo o HTML retornado
```

```python
# 2. Passar a página para o BeautifulSoup
soup = BeautifulSoup(page.content, "html.parser")
```

```python
# 3. Buscar o primeiro elemento <h1> na página
titulo = soup.find("h1")
```

```python
# 4. Exibir o elemento <h1> e seu texto
print(titulo)          # <h1 class="title is-1">Fake Python</h1>
print(titulo.get_text(strip=True))  # "Fake Python"
```

```python
# Buscando o conteúdo dos cards
import requests
from bs4 import BeautifulSoup

# 1. Fazer a requisição para a página de vagas
URL = "https://realpython.github.io/fake-jobs/"
response = requests.get(URL)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 2. Buscar o container onde estão todas as vagas
resultado = soup.find(id="ResultsContainer")
print(resultado.prettify())  # Exibe o HTML "prettificado" do container
```

```python
# 3. Encontrar todos os cards de vaga dentro desse container
cards_emprego = resultado.find_all("div", class_="card-content")
```

```python
# 4. Iterar sobre cada card para extrair título, empresa e local
vagas_emprego = []
for card_emprego in cards_emprego:
    vaga = card_emprego.find("h2", class_="title")
    empresa = card_emprego.find("h3", class_="company")
    local = card_emprego.find("p", class_="location")

    vagas_emprego.append({
        "vaga": vaga.text.strip(),
        "empresa": empresa.text.strip(),
        "local": local.text.strip()
    })
```

```python
# 5. Exibir a lista de dicionários com as informações das vagas
for vaga in vagas_emprego:
  print(vaga)
```

```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# 2.1. Aponte para o binário do Chrome estável
os.environ['GOOGLE_CHROME_BIN'] = '/usr/bin/google-chrome-stable'

# 2.2. Crie as opções de Chrome (modo headless, etc.)
chrome_options = Options()
chrome_options.add_argument('--headless')               # roda sem interface gráfica
chrome_options.add_argument('--no-sandbox')             # requerido no Colab
chrome_options.add_argument('--disable-dev-shm-usage')  # evita problemas de memória
chrome_options.binary_location = os.environ['GOOGLE_CHROME_BIN']

# 2.3. Use o webdriver-manager para baixar/instalar o ChromeDriver correto
#      O ChromeDriver (compatível com a versão do google-chrome-stable) será baixado em tempo de execução
driver_path = ChromeDriverManager().install()

# 2.4. Crie o Service apontando para o driver baixado pelo webdriver-manager
service = Service(executable_path=driver_path)

# 2.5. Finalmente, inicialize o WebDriver
driver = webdriver.Chrome(service=service, options=chrome_options)

# 2.6. Teste abrindo uma página, por exemplo o Python.org
driver.get("https://www.python.org")
print("Título da página:", driver.title)

# 2.7. Quando terminar, feche o driver
driver.quit()
```
